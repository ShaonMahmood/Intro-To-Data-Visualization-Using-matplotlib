```python
# dynamic way of doing things

data = [
    np.random.randint(1,20,10),
    np.random.randint(1,20,10),
    np.random.randint(1,20,10)
]
label_options = ["A","B","C"]
marker_options = ['s', 'o', 'd']
color_options = ['pink','b','g']
linestyle_options = ['--','-','-.']
for i in range(len(data)):
    plt.plot(data[i], label=label_options[i], marker=marker_options[i], color=color_options[i],
            linestyle=linestyle_options[i], markersize=15, lw=5, markeredgewidth=5, markerfacecolor='r')
    plt.xticks(fontsize=16); plt.yticks(fontsize=16)
    plt.grid(True)
    
    plt.legend(label_options[i], fontsize=20, loc=(1.1,0.4))
    plt.show()
    

    
```
